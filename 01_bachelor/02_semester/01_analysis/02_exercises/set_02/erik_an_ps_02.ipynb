# Problem 1

## Task

Let $X$ be a metric space and $f : [0, 1] \to X$ continuous. Show that $f([0, 1]) \subseteq X$ is path-connected.

## Solution

#+begin_quote
*Lemma 11.1.12:* Path-connectedness is preserved under continuous maps

Let $X$ and $Y$ be metric spaces. If $X$ is path-connected and $f : X \rightarrow Y$ is continuous, then $f(X)$ is path connected.

*Proof.*

Let $a_1, a_2 \in f(X)$ be arbitrary. Then there are $x_1, x_2 \in X$ such that $f(x_1) = a_1$ and $f(x_2) = a_2$. Since $X$ is connected, there is a continuous function $\gamma : \left[ 0,1 \right] \rightarrow X$ such that $\gamma(0) = x_1$ and $\gamma(1) = x_2$. The function $f \circ \gamma : \left[ 0,1 \right] \rightarrow f(X)$ is also continuous, because $f$ is assumed to be continuous. It satisfies $f \circ \gamma(0) = f(x_1) = a_1$ and $f \circ \gamma(1) = f(x_2) = a_2$.
#+end_quote

The interval $\left[ 0,1 \right]$ is path-connected, since for any $s, t \in \left[ 0,1 \right]$ the map $\gamma(u) = s + u(t - s)$ is a continuous path from $s$ to $t$. Since $f$ is continuous, $f(\left[ 0,1 \right])$ is path-connected by /Lemma 11.1.12/.

# Problem 2

## Task

Let $f : [0, 1] \to \mathbb{R}^2$ and $f_1, f_2 : [0, 1] \to \mathbb{R}$, so that
\[
f(t) = \begin{pmatrix} f_1(t) \\ f_2(t) \end{pmatrix}.
\]
We equip $\mathbb{R}^2$ with the standard metric


$$d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2}$$


(a) Prove that the coordinate map $\chi_i : \mathbb{R}^2 \to \mathbb{R}$ defined by $\chi_i(x_1, x_2) = x_i$ is continuous, for $i = 1, 2$.

(b) Prove that $f$ is continuous if and only if both $f_1$ and $f_2$ are continuous.

### (a)

## Solution

Let $\epsilon > 0$. Choose $\delta = \epsilon$. Then if $d(x,y) < \delta$,

$$|\chi_i(x) - \chi_i(y)| = |x_i - y_i| \le \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2} = d(x,y) < \delta = \epsilon$$

Hence, $\chi_i : \mathbb{R}^2 \rightarrow \mathbb{R}$ is continuous.

### (b)

## Solution

- ($\Rightarrow$) If $f$ is continuous, then $f_i = \chi_i \circ f$ is continuous as a composition of continuous maps.

- ($\Leftarrow$) Let $\epsilon > 0$. $f_1$ is continuous, so $\forall \epsilon > 0 \exists \delta_1 : |s - t| < \delta_1 \implies |f_1(s) - f_1(t)| < \frac{\epsilon}{\sqrt{2}}$. Similarly, for $f_2$, $\forall \epsilon > 0 \exists \delta_2 : |s - t| < \delta_2 \implies |f_2(s) - f_2(t)| < \frac{\epsilon}{\sqrt{2}}$.

  Define $\delta = \min(\delta_1, \delta_2)$. Thus,

  $$d(f(s), f(t)) = \sqrt{\left( f_1(s) - f_1(t) \right)^2 + \left( f_2(s) - f_2(t) \right)^2} < \sqrt{\frac{\epsilon^2}{2} + \frac{\epsilon^2}{2}} = \epsilon$$

# Problem 3

## Task

Let $S = \mathbb{R}^2 \setminus \{(0, 0)\}$ and $x = (1, 0)$ and $y = (-1, 0)$.

(a) Define a continuous path from $x$ to $y$.

(b) Plot this path in Julia.

(c) Visualize the sets

\begin{align*}
A &= \{(x, y) \in \mathbb{R}^2 : x^2 + y^2 < 1\} \\
B &= \{(-1, y) : -1 \leq y \leq 1\} \cup \{(1, y) : -1 \leq y \leq 1\} \\
\end{align*}

in Julia by sample/grid points. Determine which of the two sets is path-connected.

### (a)

## Solution

$\gamma : \left[ 0,1 \right] \rightarrow S \quad \gamma(t) = \left( \cos(\pi t), \sin(\pi t) \right)$

### (b)

```julia
using Plots

gamma(t) = (cos(π * t), sin(π * t))

ts = 0:0.01:1
xs = first.(gamma.(ts))
ys = last.(gamma.(ts))

plot(xs, ys,
     title="Continuous path",
     label="γ",
     xlabel="x", ylabel="y",
     aspect_ratio=:equal)
savefig("./assets/p_03_02.png")
```

### (c)

```julia
using Plots

ys_range = -1:0.01:1

A_xs = [x for x in ys_range for y in ys_range if x^2 + y^2 < 1]
A_ys = [y for x in ys_range for y in ys_range if x^2 + y^2 < 1]

B_xs = [fill(-1, length(ys_range)); fill(1, length(ys_range))]
B_ys = [collect(ys_range); collect(ys_range)]

opts = (xlabel="x", ylabel="y", markersize=1, aspect_ratio=:equal, legend=false)

p_1 = scatter(A_xs, A_ys; title="Set A", opts...)
p_2 = scatter(B_xs, B_ys; title="Set B", opts...)

plot(p_1, p_2)
savefig("./assets/p_03_03_01.png")
```

## Solution

Hence,

$A$ is path-connected (straight line between any two points stays inside).

$B$ is not path-connected (the two segments are disjoint).

# Problem 4

## Task

Let $f : [0, 1] \to \mathbb{R}$ be given by

$$f(x) = \frac{1}{4}(x^2 + 1)$$

(a) Show that $f$ maps $[0, 1]$ into itself and is a contraction on $[0, 1]$.

(b) Deduce that $f$ has a unique fixed point and implement the fixed point iteration in Julia.

### (a)

## Solution

- $f$ maps $[0,1]$ into itself.
  \begin{align*}
  x \in \left[ 0,1 \right] &\implies x^2 \in \left[ 0,1 \right] \\
  &\implies x^2 + 1 \in \left[ 1,2 \right] \\
  &\implies \frac{1}{4}(x^2 + 1) \in \left[ \frac{1}{4},\frac{1}{2} \right] \\
  \end{align*}

  As $\left[ \frac{1}{4},\frac{1}{2} \right] \subseteq \left[ 0,1 \right]$, $f$ maps $\left[ 0,1 \right]$ into itself.

- $f$ is a contraction on $[0,1]$. For all $x, y \in \left[ 0,1 \right]$,

  \begin{align*}
  \left| f(x) - f(y) \right| &= \frac{1}{4} \left| x^2 - y^2 \right| \\
  &= \frac{1}{4} \left| x + y \right| \left| x - y \right| \\
  &\le \frac{1}{4} \cdot 2 \cdot \left| x - y \right| \\
  &\le \frac{1}{2} \left| x - y \right|.
  \end{align*}

  Hence $f$ is a contraction with $q = \frac{1}{2}$.

### (b)

```julia
f(x) = 1/4 * (x^2 + 1)

function fixed_point_iteration(f, x_0, tol=1e-10, max_iter=100)
    x = x_0
    for n in 1:max_iter
        x_new = f(x)
        if abs(x_new - x) < tol
            println("Converged after $n iterations")
            return x_new
        end
        x = x_new
    end
    @warn "Did not converge within $max_iter iterations"
    return x
end

x_star = fixed_point_iteration(f, 0.0)
println("Fixed point: $x_star")
println("Exact value: $(2 - sqrt(3))")
```
