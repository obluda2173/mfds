## Task

Assume that your computer can execute $10^9$ operations per second. For each of the following numbers of operations $f(n)$ and the given time limits calculate the (approximal) maximal size $n$ of your problem that can be solved.

| $f(n)$      | $1\;\mu s$ | $1\;ms$ | $1\;s$ | $1\;min$ | $1\;hour$ | $1\;day$ | $1\;week$ | $30\;days$ | $1\;year$ | $1000\;years$ |
|-------------+------------+---------+--------+----------+-----------+----------+-----------+------------+-----------+---------------|
| $\lg \lg n$ |            |         |        |          |           |          |           |            |           |               |
| $\lg n$     |            |         |        |          |           |          |           |            |           |               |
| $\sqrt{n}$  |            |         |        |          |           |          |           |            |           |               |
| $n$         |            |         |        |          |           |          |           |            |           |               |
| $n \lg n$   |            |         |        |          |           |          |           |            |           |               |
| $n^2$       |            |         |        |          |           |          |           |            |           |               |
| $n^3$       |            |         |        |          |           |          |           |            |           |               |
| $n^4$       |            |         |        |          |           |          |           |            |           |               |
| $2^n$       |            |         |        |          |           |          |           |            |           |               |
| $n!$        |            |         |        |          |           |          |           |            |           |               |

```julia
import Printf: @sprintf
using Printf

const OPS_PER_SEC = 1e9

time_labels  = ["1 μs", "1 ms", "1 s", "1 min", "1 hour", "1 day", "1 week", "30 days", "1 year", "1000 years"]
time_seconds = [1e-6, 1e-3, 1.0, 60.0, 3600.0, 86400.0, 604800.0, 2.592e6, 3.1536e7, 3.1536e10]

# ── Inverse functions ──────────────────────────────────────────────

# lg lg n = ops  →  n = 2^(2^ops)
function inv_lglg(ops)
    return "2^2^$(@sprintf("%.3g", ops))"
end

# lg n = ops  →  n = 2^ops = 10^(ops * log10(2))
function inv_lg(ops)
    log10_val = ops * log10(2)
    return "10^$(@sprintf("%.3g", log10_val))"
end

# √n = ops  →  n = ops²
inv_sqrt(ops) = @sprintf("%.3g", ops^2)

# n = ops
inv_n(ops) = @sprintf("%.3g", ops)

# n lg n = ops  →  binary search
function inv_nlgn(ops)
    ops < 1 && return "0"
    lo, hi = 1.0, ops
    for _ in 1:100
        mid = (lo + hi) / 2
        val = mid * log2(mid)
        val < ops ? (lo = mid) : (hi = mid)
    end
    return @sprintf("%.3g", lo)
end

# n² = ops  →  n = √ops
inv_n2(ops) = @sprintf("%.3g", sqrt(ops))

# n³ = ops  →  n = ops^(1/3)
inv_n3(ops) = @sprintf("%.3g", ops^(1/3))

# n⁴ = ops  →  n = ops^(1/4)
inv_n4(ops) = @sprintf("%.3g", ops^(1/4))

# 2ⁿ = ops  →  n = floor(log₂(ops))
function inv_2n(ops)
    ops < 1 && return "0"
    return @sprintf("%.3g", floor(log2(ops)))
end

# n! = ops  →  largest n where n! ≤ ops
function inv_nfact(ops)
    ops < 1 && return "0"
    n = BigInt(1)
    fact = BigInt(1)
    ops_big = BigInt(floor(ops))

    while true
        next_fact = fact * (n + 1)
        if next_fact > ops_big
            break
        end
        n += 1
        fact = next_fact
    end
    return string(n)
end

# ── Build table ────────────────────────────────────────────────────

func_labels = ["lg lg n", "lg n", "√n", "n", "n lg n", "n²", "n³", "n⁴", "2ⁿ", "n!"]
inv_funcs   = [inv_lglg, inv_lg, inv_sqrt, inv_n, inv_nlgn, inv_n2, inv_n3, inv_n4, inv_2n, inv_nfact]

results = Matrix{String}(undef, length(func_labels), length(time_labels))

for (j, t) in enumerate(time_seconds)
    ops = t * OPS_PER_SEC
    for (i, f) in enumerate(inv_funcs)
        results[i, j] = f(ops)
    end
end

# ── Pretty print (Org/Markdown Format) ─────────────────────────────

header = "| \$f(n)\$ | " * join(time_labels, " | ") * " |"
separator = "|---|$(join(fill("---", length(time_labels)), "|"))|"

println(header)
println(separator)
for i in 1:length(func_labels)
    row = "| \$$(func_labels[i])\$ | " * join(results[i, :], " | ") * " |"
    println(row)
end
```

# Problem 2

## Task

Consider the algorithm SUM ARRAY from page 25 of the lecture book:

1: procedure SUM ARRAY
2:   sum = 0
3:   for i = 1 to n do
4:     sum = sum + A[i]
5:   end for
6: end procedure

It should solve the problem to sum the elements of the array $A$. Prove that the algorithm is correct by using a loop invariant.

## Solution

*Loop invariant:* At the start of each iteration with index $i$, $sum = A[1] + A[2] + \cdots + A[i - 1]$.

*Initialization (i = 1):* $sum = 0$, which equals the empty sum $A[1] + \cdots + A[0]$ (a sum of zero terms). This holds.

*Maintenance:* Assume at the start of iteration $i$ we have $sum = A[1] + \cdots + A \left[ i - 1 \right]$. The loop body executes $sum = sum + A[i]$, so after the body, $sum = A[1] + \cdots + A[i]$. When $i$ increments to $i+1$, we're at the start of the next iteration and $sum = A[1] + \cdots + A[i + 1 - 1 ]$. The invariant holds for the next iteration.

*Termination:* The loop ends when $i = n + 1$. Substituting into the invariant: $sum = A[1] + \cdots + A[n]$, which is exactly what we wanted to prove.

# Problem 3

## Task

We define the searching problem as follows: Input: A sorted sequence of $n$ numbers $(a_1, a_2, \ldots, a_n)$ with $a_i < a_{i+1}$ for all $i < n$ provided in array $A$, a value $x$. Output: An index $i$ such that $x = A[i]$ and if $x$ does not appear in $A$, return the largest index $i$ such that $A[i] < x$. If no such index exists, return $0$. Write pseudocode for linear search that scans the array from the beginning to the end to determine $i$. Implement the procedure in Julia.

```julia
function linear_search(array, x)
    n = length(array)
    i = 1
    while (i <= n) && (array[i] <= x)
        i += 1
    end
    return i - 1
end

function main()
    A = [1, 2, 3, 4, 6, 7, 8, 9]
    println(linear_search(A, 4))
    println(linear_search(A, 5))
    println(linear_search(A, 0))
    println(linear_search(A, 10))
end

main()
```

# Problem 4

## Task

Using a loop invariant, prove that your algorithm from Exercise 3 is correct. Make sure that your loop invariant fulfills the three necessary properties. Give an asymptotic running time using the Landau symbols for the worst-case.

## Solution

*Loop invariant:* $\forall j : 1 \le j < i, A \left[ j \right] \le x$

*Initialization (i = 1):* The range $j = 1, \cdots, 0$ is empty, so the statement "for all $j < 1$, $A[j] \le x$" holds vacuously.

*Maintenance:* Assume the invariant holds at the start of iteration $i$, meaning $A[1], \ldots, A[i-1]$ are all $\le x$. The loop condition checks that $i \le n$ and $A[i] \le x$. Since we enter the body, we know $A[i] \le x$. After incrementing $i$ to $i+1$, we now know $A[1], \ldots, A[(i+1)-1]$ are all $\le x$. The invariant holds for the next iteration.

*Termination:* The loop exits when either $i > n$ or $A[i] > x$. In both cases the invariant tells us $A[1], \ldots, A[i-1]$ are all $\le x$. Since the array is sorted, $i-1$ is the largest index with $A[i-1] \le x$. If $A[i-1] = x$, then $i-1$ is the index of $x$. If $A[i-1] < x$ (or $i = 1$), then $x$ is not in the array and $i-1$ is the largest index with $A[i-1] < x$ (or $0$ if no such index exists). This matches the required output.

*Worst-case running time:* The loop executes at most $n$ iterations (when all elements are $\le x$), and each iteration does constant work, so the worst-case running time is $\mathcal{O}(n)$.

# Problem 5

## Task

Let an array $A[1 : n]$ of numbers be given. Consider the following algorithm for solving the sorting problem: Find the smallest element of $A[1 : n]$ and then exchange it with the first element. Afterwards, find the smallest element in $A[2 : n]$ and exchange it with the second element, and so on. This algorithm is called selection sort. Write pseudocode for this algorithm and implement it in Julia.

```julia
function selection_sort(array)
    n = length(array)

    for i in 1:n
        j = argmin(array[i:end]) + i - 1
        array[i], array[j] = array[j], array[i]
    end
    return array
end

function main()
    array = [9, 8, 7, 6, 5, 4, 3, 2, 1]
    println(selection_sort(array))
end

main()
```

# Problem 6

## Task

Take the algorithm selection sort from Exercise 5 and find a loop invariant for that algorithm. Prove your assumption.

## Solution

*Loop invariant:* At the start of each iteration with index $i$, the subarray $A[1], \ldots, A[i-1]$ is sorted and contains the $i-1$ smallest elements of the original array.

*Initialization (i = 1):* The subarray $A[1], \cdots, A[0]$ is empty. An empty subarray is trivially sorted and trivially contains the $0$ smallest elements.

*Maintenance:* Assume at the start of iteration $i$, $A[1], \cdots, A[i-1]$ is sorted and contains the $i-1$ smallest elements. The algorithm finds the minimum of $A[i], \cdots, A[n]$ and swaps it into position $i$. Since $A[1], \cdots, A[i-1]$ already contains the $i-1$ smallest elements, the minimum of the remaining subarray is the $i$-th smallest element overall. After the swap, $A[i]$ holds this value, so $A[1], \cdots, A[i]$ is sorted and contains the $i$ smallest elements. When $i$ increments, the invariant holds for the next iteration.

*Termination:* The loop ends when $i = n + 1$. The invariant gives us that $A[1], \cdots, A[n]$ is sorted and contains the $n$ smallest elements, which means the entire array is sorted.

# Problem 7

## Task

Analyse your algorithm from Exercise 5 and determine its best- and worst-case running times. Provide asymptotic running times using the Landau symbols ($O$, $\Omega$, or $\Theta$), whichever fit best.

## Solution

The outer loop runs from $i = 1$ to $n$. In each iteration, finding the minimum
of $A[i \ldots n]$ requires $n - i$ comparisons, and the swap is constant time.
This work is the same regardless of the input.

Total comparisons:
$$\sum_{i=1}^{n}(n - i) = \frac{n(n-1)}{2} = \frac{n^2}{2} - \frac{n}{2}$$

*Upper bound ($O$):* For $n \geq 1$, we have $\frac{n^2}{2} - \frac{n}{2} \leq \frac{n^2}{2} \leq n^2$.
Choose $c = 1, \, n_0 = 1$. So $f(n) = O(n^2)$.

*Lower bound ($\Omega$):* For $n \geq 2$, we have $\frac{n}{2} \leq \frac{n^2}{4}$,
so $\frac{n^2}{2} - \frac{n}{2} \geq \frac{n^2}{2} - \frac{n^2}{4} = \frac{n^2}{4}$.
Choose $c = \frac{1}{4}, \, n_0 = 2$. So $f(n) = \Omega(n^2)$.

Since $f(n) = O(n^2)$ and $f(n) = \Omega(n^2)$, we conclude $f(n) = \Theta(n^2)$.

Since selection sort always performs the same number of comparisons regardless
of the input, the best-case and worst-case running times are both $\Theta(n^2)$.

# Problem 8

## Task

Use mathematical induction to show that when $n = 2^k$ for $k \geq 1$, the solution of the recurrence

$T(n) = \begin{cases} 2 & n = 2 \\ 2T\left(\frac{n}{2}\right) + n & n > 2 \end{cases}$

is $T(n) = n \lg n$.

## Solution

- Base case

  $k = 1$, so $n = 2$. Hence $T(2) = 2 \cdot \lg(2) = 2 \cdot 1 = 2$. Base holds.

- Inductive step

  Assume $T(2^k) = 2^k \cdot k$ (since $\lg(2^k) = k$).

  Now show that $T(2^{k+1}) = 2^{k+1} \cdot (k+1)$.

  Hence,

  \begin{align*}
  T(2^{k+1}) &= 2 \cdot T(\frac{2^{k+1}}{2}) + 2^{k+1} \\
             &= 2 \cdot T(2^{k}) + 2^{k+1} \\
             &= 2 \cdot (2^k \cdot k) + 2^{k+1} \\
             &= 2^{k+1} \cdot k + 2^{k+1} \\
             &= 2^{k+1}(k + 1) \\
  \end{align*}

Hence, $T(n) = n \lg(n)$

# Problem 9

## Task

(a) Express $123n^2 + 345n - 148$ as $\Theta$ of a simple expression.

(b) Is $2^{n+1} = O(2^n)$?

(c) Is $2^{2n} = O(2^n)$?

(d) Is $2^n = O(2^{2n})$?

(e) Is $2^{n+1} = o(2^n)$?

(f) Is $2^n = o(2^{2n})$?

### (a)

## Solution

- Upper bound

  $$123n^2 + 345n - 148 \le 123n^2 + 345n^2 = 468n^2$$

  Choose $c = 468, n_0 = 1$. So, $f(n) = O(n^2)$

- Lower bound

  $$123n^2 + 345n - 148 \ge 123n^2 + 345 - 148 \ge 123n^2$$

  Choose $c = 123, n_0 = 1$. So $f(n) = \Omega(n^2)$
 
Since $f(n) = O(n^2)$ and $f(n) = \Omega(n^2)$, we conclude $f(n) = \Theta(n^2)$.

### (b)

## Solution

$$2^{n+1} = 2 \cdot 2^n$$

Choose $c = 2, n \ge 1$. $f(n) = O(2^n)$

### (c)

## Solution

Assume $2^{2n} = O(2^n)$. Hence,

\begin{align*}
2^{2n} &\le c \cdot 2^n \\
2^n \cdot 2^n &\le c \cdot 2^n \\
2^n &\le c \\
\end{align*}

But $2^n$ grows without a bound, hence, no constant $c$ can satisfy this for all sufficiently large $n$.

Thus,

$$2^{2n} \neq O(2^n)$$

### (d)

## Solution

Assume $2^n = O(2^{2n})$. Let $c = 1$, hence,

\begin{align*}
2^n &\le 1 \cdot 2^n \cdot 2^n \\
1 &\le 2^n \\
\end{align*}

Which is true $\forall n \ge 1$.

### (e)

## Solution

$2^{n+1} = o(2^n)$ iff:

\begin{align*}
\lim_{n\rightarrow\infty} \dfrac{2^{n+1}}{2^n} &= 0 \\
\lim_{n\rightarrow\infty} \dfrac{2 \cdot 2^n}{2^n} &= 0 \\
2 \cdot \lim_{n\rightarrow\infty} \dfrac{2^n}{2^n} &= 0 \\
2 \cdot 1 &= 0 \\
2 &\neq 0 \\
\end{align*}

Hence,

$$2^{n+1} \neq o(2^n)$$

### (f)

## Solution

$2^n = o(2^{2n})$ iff:

\begin{align*}
\lim_{n\rightarrow\infty} \dfrac{2^n}{2^{2n}} &= 0 \\
\lim_{n\rightarrow\infty} \dfrac{2^n}{2^n \cdot 2^n} &= 0 \\
\lim_{n\rightarrow\infty} \dfrac{1}{2^n} &= 0 \\
0 &= 0 \\
\end{align*}

Hence,

$$2^n = o(2^{2n})$$

# Problem 10

## Task

Consider sequences $f(n)$, $g(n)$ and $h(n)$. Assume that $f(n) = \Theta(g(n))$ and that $g(n) = \Theta(h(n))$. Show that then $f(n) = \Theta(h(n))$.

## Solution

Assume $f(n) = \Theta(g(n)), g(n) = \Theta(h(n))$. Show $f(n) = \Theta(h(n))$.

From $f(n) = \Theta(g(n))$: $\exists c_1, c_2 > 0, \, n_1 : c_1 \cdot g(n) \le f(n) \le c_2 \cdot g(n), \forall n \ge n_1$.

From $g(n) = \Theta(h(n))$: $\exists c_3, c_4 > 0, \, n_2 : c_3 \cdot h(n) \le g(n) \le c_4 \cdot h(n), \forall n \ge n_2$.

Let $n_0 = \max(n_1, n_2)$. Then for all $n \ge n_0$:

\begin{align*}
c_1 \cdot g(n) \le &f(n) \le c_2 \cdot g(n) \\
c_3 \cdot h(n) \le &g(n) \le c_4 \cdot h(n) \\
c_1 \cdot c_3 \cdot h(n) \le c_1 \cdot g(n) \le &f(n) \le c_2 \cdot g(n) \le c_2 \cdot c_4 \cdot h(n) \\
\end{align*}

Let $c_5 = c_1 \cdot c_3, \, c_6 = c_2 \cdot c_4$. Then:

$$c_5 \cdot h(n) \le f(n) \le c_6 \cdot h(n), \quad \forall n \ge n_0$$

Thus, $f(n) = \Theta(h(n))$. $\blacksquare$
