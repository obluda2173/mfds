## Task
> (Programming, Julia) Implement two Julia functions `forward_substitution(L,b)` and `backward_substitution(U,y)` that solve $Ly=b$ and $Ux=y$ for triangular matrices of compatible sizes. Use your code to solve the system from this exercise and verify that your result agrees with Julia's A \ b.

## Solution

```julia
is_compatible(m_1, m_2) = size(m_1, 2) == size(m_2, 1)

multiply(row, column) = row' * column

function forward_substitution(L, b)
    if !is_compatible(L, b)
        error("Matrix Dimensions Mismatch.")
    end

    n = size(L, 1)

    x = zeros(Float64, n)

    for i in 1:n
        sum_val = b[i]
        for j in 1:(i-1)
            sum_val -= L[i, j] * x[j]
        end
        x[i] = sum_val / L[i, i]
    end
    return x
end

function backward_substitution(U, y)
    if !is_compatible(U, y)
        error("Matrix Dimensions Mismatch.")
    end

    n = size(U, 1)

    x = zeros(Float64, n)

    for i in n:-1:1
        sum_val = y[i]
        for j in (i+1):n
            sum_val -= U[i, j] * x[j]
        end
        x[i] = sum_val / U[i, i]
    end
    return x
end

L = [ 1  0  0;
      2  1  0;
     -1  2  1]

U = [ 2  1 -1;
      0  3 -1;
      0  0 -1]

b = [ 1;
      7;
      3]

y = forward_substitution(L, b)
x = backward_substitution(U, y)

println("Intermediate vector y: ", y)
println("Final solution x: ", x)
```
