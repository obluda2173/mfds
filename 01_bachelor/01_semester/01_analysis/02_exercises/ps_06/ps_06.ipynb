### Problem 2

**Task:**
Approximate the following integrals in Julia by choosing the appropriate equidistant partition of the respective intervals and implementing $L(f,\mathcal{P}_{n})$ and $U(f,\mathcal{P}_{n})$:
(a) $\int_{1}^{3}x^{2}dx$
(b) $\int_{0}^{\log 2}e^{x}dx$
(c) $\int_{0}^{\pi}\sin x dx$
(d) $\int_{1}^{e}\frac{1}{x}dx$

```julia
function L(f, P)
    area = 0.0
    for i in 1:length(P)-1
        h = min(f(P[i]), f(P[i+1]))

        w = P[i+1] - P[i]

        area += h * w
    end
    return area
end

function U(f, P)
    area = 0.0
    for i in 1:length(P)-1
        h = max(f(P[i]), f(P[i+1]))

        w = P[i+1] - P[i]

        area += h * w
    end
    return area
end

function exercise()
    n = 1000

    # each item has: label, function, start(a), end(b), and exact answer
    data = [
        (label="(a) x^2",     f = x -> x^2,     a = 1,  b = 3,       exact = 26/3),
        (label="(b) e^x",     f = x -> exp(x),  a = 0,  b = log(2),  exact = 1   ),
        (label="(c) sin(x)",  f = x -> sin(x),  a = 0,  b = pi,      exact = 2   ),
        (label="(d) 1/x",     f = x -> 1/x,     a = 1,  b = exp(1),  exact = 1   )
    ]

    println("Approximating integrals with n = $n partitions:\n")
    println(rpad("Problem", 15), rpad("Lower Sum (L)", 18), rpad("Upper Sum (U)", 18), "Exact")
    println("-"^65)

    for i in data
        P_n = range(i.a, i.b, length=n+1)

        val_L = L(i.f, P_n)
        val_U = U(i.f, P_n)

        print(rpad(i.label, 15))
        print(rpad(round(val_L, digits=5), 18))
        print(rpad(round(val_U, digits=5), 18))
        println(round(i.exact, digits=5))
    end

end

exercise()
```

### Problem 3

**Task:**
Implement the trapezoidal rule in Julia for equidistant partitions and compare the approximation with the exact value for the following integrals:
(a) $\int_{0}^{1}xe^{x}dx$
(b) $\int_{0}^{\pi}x \sin x dx$
(c) $\int_{1}^{e}\log x dx$
(d) $\int_{0}^{1}x \log x dx$
(e) $\int_{0}^{1}x^{2}e^{x}dx$

```julia
using Printf

function T(f, P)
    area = 0.0
    for i in 1:length(P)-1
        h = (f(P[i]) + f(P[i+1])) / 2
        w = P[i+1] - P[i]
        area += h * w
    end
    return area
end

function exercise()
    n = 1000

    safe_xlogx(x) = (x == 0) ? 0.0 : x * log(x)

    data = [
        (label="(a) x e^x",     f = x -> x * exp(x),    a = 0.0, b = 1.0,     exact = 1.0     ),
        (label="(b) x sin(x)",  f = x -> x * sin(x),    a = 0.0, b = pi,      exact = pi      ),
        (label="(c) log(x)",    f = x -> log(x),        a = 1.0, b = exp(1),  exact = 1.0     ),
        (label="(d) x log(x)",  f = safe_xlogx,         a = 0.0, b = 1.0,     exact = -0.25   ),
        (label="(e) x^2 e^x",   f = x -> x^2 * exp(x),  a = 0.0, b = 1.0,     exact = exp(1)-2)
    ]

    println("Approximating integrals using Trapezoidal Rule (n = $n):\n")
    println(rpad("Problem", 16), rpad("Approx (T)", 18), rpad("Exact", 18), "Diff")
    println("-"^65)

    for i in data
        P_n = range(i.a, i.b, length=n+1)

        val_T = T(i.f, P_n)
        diff = abs(val_T - i.exact)

        print(rpad(i.label, 16))
        print(rpad(round(val_T, digits=6), 18))
        print(rpad(round(i.exact, digits=6), 18))
        @printf "%.2e\n" diff
    end
end

exercise()
```

### Problem 4

**Task:**
Implement the mid-point rule in Julia for equidistant partitions and compare the approximation with the exact value.
(a) $\int_{0}^{1}2x e^{x^{2}}dx$
(b) $\int_{0}^{\pi/2}\sin(x)\cos(x)dx$
(c) $\int_{1}^{2}\frac{2x}{1+x^{2}}dx$
(d) $\int_{0}^{\log 2}e^{2x}dx$
(e) $\int_{0}^{1}\frac{1}{\sqrt{1-x^{2}}}dx$

```julia
using Printf

function M(f, P)
    area = 0.0
    for i in 1:length(P)-1
        mid = (P[i+1] + P[i]) / 2
        h = f(mid)
        w = P[i+1] - P[i]
        area += h * w
    end
    return area
end

function exercise()
    n = 1000

    data = [
        (label="(a) 2x e^(x^2)",   f = x -> 2x * exp(x^2),     a = 0.0, b = 1.0,    exact = exp(1)-1),
        (label="(b) sin(x)cos(x)", f = x -> sin(x)*cos(x),     a = 0.0, b = pi/2,   exact = 0.5),
        (label="(c) 2x/(1+x^2)",   f = x -> (2x)/(1+x^2),      a = 1.0, b = 2.0,    exact = log(5) - log(2)),
        (label="(d) e^(2x)",       f = x -> exp(2x),           a = 0.0, b = log(2), exact = 1.5),
        (label="(e) 1/sqrt(1-x^2)",f = x -> 1/sqrt(1-x^2),     a = 0.0, b = 1.0,    exact = pi/2)
    ]

    println("Approximating integrals using Midpoint Rule (n = $n):\n")
    println(rpad("Problem", 18), rpad("Approx (M)", 18), rpad("Exact", 18), "Diff")
    println("-"^65)

    for i in data
        P_n = range(i.a, i.b, length=n+1)

        val_M = M(i.f, P_n)
        diff = abs(val_M - i.exact)

        print(rpad(i.label, 18))
        print(rpad(round(val_M, digits=6), 18))
        print(rpad(round(i.exact, digits=6), 18))
        @printf "%.2e\n" diff
    end
end

exercise()
```
